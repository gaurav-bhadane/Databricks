### Row Filtering

```python
df_filtered = df.filter((df.release_year >= 2000) & (df.release_year <= 2010))
display(df_filtered)
```

```python
from pyspark.sql.functions import col
df_filtered = df.filter((col("release_year") >= 2000) & (col("release_year") <= 2010))
display(df_filtered)
```

```python
df_marvel=df.filter(col("studio")=="Marvel Studios")
display(df_marvel)
```

```python
display(df.select("industry").distinct())
```

```python
df=df.withColumn("profit",col("revenue")-col("budget"))
display(df)
```
