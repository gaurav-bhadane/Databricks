## Spark Execution

```python
df=spark.table("workspace.default.movies")
display(df)
```

```python
from pyspark.sql.functions import col

df_narrow=df.select("title","studio","imdb_rating").filter(col("release_year")>2010)

display(df_narrow)
```

### Print Logical and Physical Plan

```python
df_narrow.explain("extended")
```

### Print only Physical Plan

```python
df_narrow.explain("formatted")
```
