```python
%sql
select distinct studio from movies
```

```python
repartitionByKey=df.repartition(6,"studio")
repartitionByKey.explain("formatted")
```

```python
rep_round_robin=df.repartition(6)
rep_round_robin.explain("formatted")
```

```python
out_path="/Volumes/workspace/default/repartion_demo/repartition_6"
rep_round_robin.write.mode("overwrite").parquet(out_path)
display(dbutils.fs.ls(out_path))
```

- repartition(n, col) → Groups same keys together
- repartition(n) → Distributes rows evenly
- Distinct values ≠ partitions

```python
from pyspark.sql import Window, functions as F

# One-time shuffle:
base = df.repartition(6, "studio")

# Reuse partitioning on the same detailed rows:
agg    = base.groupBy("studio").agg(F.avg(F.col("revenue").cast("double")).alias("avg_revenue"))

display(agg)
```
